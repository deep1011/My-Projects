```python
### Analysis 1: Which was the best month for sales? 
df=df[df['Order Date'].str[0:2] != 'Or']
```

```python
df['Month']=df['Order Date'].str[0:2]
df['Month'].astype('int32')
```

```python
df.head()
```

```python
look_up={'01':'Jan', '02': 'Feb', '03': 'Mar', '04': 'Apr', '05': 'May',
           '06': 'Jun', '07': 'Jul', '08': 'Aug', '09': 'Sep', '10': 'Oct', '11': 'Nov', '12': 'Dec'}
df['Month']=df['Month'].apply(lambda x: look_up[x])
```

```python
df.head()
```

```python
df['Quantity Ordered']=df['Quantity Ordered'].astype('float')
df['Price Each']=df['Price Each'].astype('float')
```

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
```

```python
df.head()
```

```python
highest_sales = df.groupby('Month').sum()
```

```python
highest_sales.head(12)
```

```python
plt.figure(figsize = (12, 5))
plt.bar(highest_sales.index,highest_sales['Sales'], color ='red',
        width = 0.4)
plt.xlabel('Month Data')
plt.ylabel('Sales Data')
plt.show()
```

```python
# Analysis 2:  Which city sold the most product?
df['City']=df['Purchase Address'].apply(lambda x: x.split(',')[1])
```

```python
df.head()
```

```python
city_sold = df.groupby('City').sum()
```

```python
city_sold.head(9)
```

```python
plt.figure(figsize = (12, 5))
plt.bar(city_sold.index,city_sold['Quantity Ordered'],color ='blue',
        width = 0.4)
plt.xlabel('city Data')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
# Analysis 3: What time should we display advertisements to maximize likelihood of customer’s buying products?
df['Time'] = df['Order Date'].str[8:11]
```

```python
df.head()
```

```python
time_df = df.groupby('Time').sum()
```

```python
time_df.head()
```

```python
plt.show()
plt.plot(time_df.index,time_df['Quantity Ordered'])
plt.show()
```

```python
# Analysis 4: Which Products are most often sold together?
df.head()
```

```python
duplicated_data = df[df['Order ID'].duplicated(keep=False)]
```

```python
duplicated_data.head()
```

```python
duplicated_data['Product_bundle']=duplicated_data.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
duplicated_data.head()
```

```python
duplicated_data=duplicated_data[['Order ID','Product_bundle']].drop_duplicates()
```

```python
duplicated_data.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
count=Counter()
for row in duplicated_data['Product_bundle']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,3)))
count.most_common(10)
```

```python
# Analysis 5: Which product sold the most? 
product_max = df.groupby('Product').sum()
```

```python
product_max
```

```python
plt.figure(figsize = (12, 5))
plt.bar(product_max.index,product_max['Quantity Ordered'],color ='orange',
        width = 0.4)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure()
fig,ax1=plt.subplots()
ax2=ax1.twinx()
ax1.bar(product_max.index,product_max['Quantity Ordered'],color ='blue',
        width = 0.4)
plt.xticks(rotation=45)
ax2.plot(product_max.index,product_max['Sales'],color='r')
plt.show()
```
